```python
import matplotlib as mlib
import matplotlib.pyplot as plt
import numpy as np
from numpy import genfromtxt
```

```python
cross200mkm_scalar = np.array(genfromtxt('cross-sections_ctau-100mkm_scalar.dat'))
cross200mkm_pseudo = np.array(genfromtxt('cross-sections_ctau-100mkm_scalar.dat'))
cross1cm_scalar = np.array(genfromtxt('cross-sections_ctau-1cm_scalar.dat'))
cross1cm_pseudo = np.array(genfromtxt('cross-sections_ctau-1cm_pseudo.dat'))
cross_charged = np.array(genfromtxt('cross-sections_charged.dat'))
```

```python
#Plotting settings
font = {
        'weight' : 'normal',
        'size'   : 16}

mlib.rc('font', **font)
```

```python
font = {
        'weight' : 'normal',
        'size'   : 16}

mlib.rc('font', **font)

plt.figure(figsize=(9, 7))


plt.plot(cross200mkm_scalar[:,0],cross200mkm_scalar[:,1], c="blue",linestyle='-',\
            label='$c\\tau \sim 100\ [\mu m]$')
plt.plot(cross1cm_scalar[:,0],cross1cm_scalar[:,1], c="orange",linestyle='-',\
            label='$c\\tau \sim 1\ [cm]$')

#plt.ylim( (2.2, 8.3))
#plt.xlim((0.015,150))
plt.yscale("log")
plt.legend(loc='upper right')

#plt.title("scalar scenario")
plt.xlabel('$m_T$ [GeV]')
plt.ylabel('$\\sigma$ [pb]')
plt.savefig('cross_sections_scalar.png')
```

```python
font = {
        'weight' : 'normal',
        'size'   : 15}

mlib.rc('font', **font)

plt.figure(figsize=(7.5, 5))


#plt.plot(cross200mkm_scalar[:,0],cross200mkm_scalar[:,1], c="blue",linestyle='-',\
#            label='whole brahch, $c\\tau \sim 100\ [\mu m]$')
plt.plot(cross_charged[:,0],cross_charged[:,1], c="firebrick",linewidth=3,\
            label='$\\chi_\\pm$ production')

plt.ylim( (0.001, 10))
#plt.xlim((0.015,150))
plt.yscale("log")
#plt.legend(loc='upper right')

plt.text(350, 0.2, "$p p \\rightarrow \\chi^+ \\chi^- $", fontsize=16)

#plt.title("scalar scenario")
plt.xlabel('$m_c$ [GeV]')
plt.ylabel('$\\sigma$ [pb]')
plt.savefig('cross_sections_whole_and_charged.png')
```
